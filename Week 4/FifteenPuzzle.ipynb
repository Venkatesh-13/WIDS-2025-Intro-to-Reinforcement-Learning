```python
#Defining states
from itertools import permutations

row1states = []
row2states = []
row3states = []
goal1 = tuple([_ for _ in range(1, 5)])
goal2 = tuple([_ for _ in range(1,9)])
goal3 = tuple([_ for _ in range(1, 17)])

for i in range(16):
    for j in range(16):
        for k in range(16):
            for l in range(16):
                a = [i-j, j-k, k-l, i-k, j-l, i-l]
                if 0 not in a:
                    for m in range(16):
                        if(m!=i and m!=j and m!=k and m!=l):
                            tmp = [0]*16
                            tmp[i]=1
                            tmp[j]=2
                            tmp[k]=3
                            tmp[l]=4
                            tmp[m]=16
                            row1states.append(tuple(tmp))
                        
for i in range(4, 16):
    for j in range(4, 16):
        for k in range(4, 16):
            for l in range(4, 16):
                a = [i-j, j-k, k-l, i-k, j-l, i-l]
                if 0 not in a:
                    for m in range(4, 16):
                        if(m!=i and m!=j and m!=k and m!=l):
                            tmp = [1,2,3,4]+[0]*12
                            tmp[i]=5
                            tmp[j]=6
                            tmp[k]=7
                            tmp[l]=8
                            tmp[m]=16
                            row2states.append(tuple(tmp))

for i in permutations([_ for _ in range(8, 16)]):
    tmp = [_ for _ in range(1,9)]+[0]*8
    for j in range(7):
        tmp[i[j]]=j+9
    tmp[i[-1]]=16
    row3states.append(tuple(tmp))   
```

```python
#Making MDPs
row1mdp = {_:{} for _ in row1states}
row2mdp = {_:{} for _ in row2states}
row3mdp = {_:{} for _ in row3states}

#Row 1 MDP Filling
for state in row1mdp:
    actions = ['u', 'd', 'l', 'r']
    transitions = {}
    stl = list(state)
    stg = tuple(state[0:4])
    for act in actions:
        next_state = list(state).copy()
        hole = next_state.index(16)
        if stg == goal1:
            transitions[act]=[(1, state, 0, False)] #(prob, next, reward, is next a reward)
            continue
        if act=='l':
            if hole%4 == 0:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-1] = next_state[hole-1], next_state[hole]
        elif act=='r':
            if hole%4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+1] = next_state[hole+1], next_state[hole]
        elif act=='u':
            if hole//4 == 0:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-4] = next_state[hole-4], next_state[hole]
        elif act=='d':
            if hole//4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+4] = next_state[hole+4], next_state[hole]
        corr = 0
        for i in range(4):
            if(next_state[i]==goal1[i]):
                corr+=1
        rew = -2+0.5*corr
        if next_state[0:4]==list(goal1):
            rew = 67
        transitions[act]=[(1, tuple(next_state), rew, rew==1)]
    row1mdp[state]=transitions

#Row 2 MDP Filling
for state in row2mdp:
    actions = ['u', 'd', 'l', 'r']
    transitions = {}
    stl = list(state)
    stg = tuple(state[0:8])
    for act in actions:
        next_state = list(state).copy()
        hole = next_state.index(16)
        if stg == goal2:
            transitions[act]=[(1, state, 0, False)] #(prob, next, reward, is next a reward)
            continue
        if act=='l':
            if hole%4 == 0:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-1] = next_state[hole-1], next_state[hole]
        elif act=='r':
            if hole%4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+1] = next_state[hole+1], next_state[hole]
        elif act=='u':
            if hole//4 <= 1:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-4] = next_state[hole-4], next_state[hole]
        elif act=='d':
            if hole//4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+4] = next_state[hole+4], next_state[hole]
        corr = 0
        for i in range(8):
            if(next_state[i]==goal2[i]):
                corr+=1
        rew = -5+0.5*corr
        if next_state[0:8]==list(goal2):
            rew = 140
        transitions[act]=[(1, tuple(next_state), rew, rew==1)]
    row2mdp[state]=transitions

#Row 3 MDP Filling
for state in row3mdp:
    actions = ['u', 'd', 'l', 'r']
    transitions = {}
    stl = list(state)
    stg = tuple(state[0:16])
    for act in actions:
        next_state = list(state).copy()
        hole = next_state.index(16)
        if stg == goal3:
            transitions[act]=[(1, state, 0, False)] #(prob, next, reward, is next a reward)
            continue
        if act=='l':
            if hole%4 == 0:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-1] = next_state[hole-1], next_state[hole]
        elif act=='r':
            if hole%4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+1] = next_state[hole+1], next_state[hole]
        elif act=='u':
            if hole//4 <= 2:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole-4] = next_state[hole-4], next_state[hole]
        elif act=='d':
            if hole//4 == 3:
                transitions[act]=[(1, tuple(next_state), -67, False)]
                continue
            next_state[hole], next_state[hole+4] = next_state[hole+4], next_state[hole]
        corr = 0
        for i in range(16):
            if(next_state[i]==goal3[i]):
                corr+=1
        rew = -8+0.5*corr
        if next_state==list(goal3):
            rew = 250
        transitions[act]=[(1, tuple(next_state), rew, rew==1)]
    row3mdp[state]=transitions
```

```python
def policyiteration(states, goal, mdp, num, gamma):
    actions=['u', 'd', 'r', 'l']
    v = {} #ValueFunction
    newv = {}
    oldpolicy = {} #DeterministicPolicy
    newpolicy = {}
    for _ in states:
        v[_]=10.0
        newv[_] = 0.0
        oldpolicy[_]='u'
        newpolicy[_] = 'd'
        if(goal==_[:num]):
            v[_]=1
            newv[_]=1
    smol = 1e-4

    while(newpolicy != oldpolicy):
        # for _ in states:
        #     v[_]=10.0
        #     newv[_] = 0.0
        #     if(goal==_[:num]):
        #         v[_]=1
        #         newv[_]=1
        oldpolicy=newpolicy.copy()
        #PolicyEvaluation
        #while(max(abs(newv[tuple(s)]-v[tuple(s)]) for s in states) > smol):
        while True:
            delta = 0
            v = newv.copy()
            for state in states:
                if(list(state)[0:num]==goal):
                    continue
                x = 0
                l = mdp[state][newpolicy[state]]
                for i in l:
                    x+=i[0]*(i[2]+gamma*v[i[1]]* (i[3] == False))
                newv[state]=x
                delta = max(delta, abs(newv[state]-v[state]))
            if delta<smol:
                break

        #PolicyImprovement
        for state in states:
            if(list(state)[0:num]==goal):
                continue
            besta = 'u'
            bestexpre = -float('inf')
            for action in actions:
                expre = 0
                l = mdp[state][action]
                for i in l:
                    expre+=i[0]*(i[2]+gamma*newv[i[1]]* (i[3] == False))
                if(expre>bestexpre):
                    besta = action
                    bestexpre = expre
            newpolicy[state]=besta
    
    return newpolicy

def valueiteration(states, goal, mdp, num, gamma):
    actions=['u', 'd', 'r', 'l']
    v = {} #ValueFunction
    newv = {}
    oldpolicy = {} #DeterministicPolicy
    newpolicy = {}
    for _ in states:
        v[_]=-10.0
        newv[_] = 0.0
        oldpolicy[_]='u'
        newpolicy[_] = 'd'
        if(goal==_[:num]):
            v[_]=1
            newv[_]=1
    smol = 1e-2

    it = 0
    while True:
        it+=1
        oldpolicy = newpolicy.copy()
        v = newv.copy()
        delt = 0
        #BellmanOptimality Update
        for state in states:
            if(list(state)[0:num]==goal):
                continue
            curr = -float('inf')
            x = 0
            for action in actions:
                a1 = mdp[state][action][0]
                s = (a1[0]*(a1[2]+gamma*v[a1[1]]))
                if(s>curr):
                    curr = s
                    x = action
            newv[state] = curr
            delt = max(delt, abs(newv[state]-v[state]))
        
        if(delt<smol):
            for state in states:
                if(list(state)[0:num]==goal):
                    continue
                curr = -float('inf')
                x = 0
                for action in actions:
                    a1 = mdp[state][action][0]
                    s = (a1[0]*(a1[2]+gamma*newv[a1[1]]))
                    if(s>curr):
                        curr = s
                        x = action
                newpolicy[state] = x
            break
    return newpolicy

row1policy = (valueiteration(row1states, goal1, row1mdp, 4, 0.9))
row2policy = (valueiteration(row2states, goal2, row2mdp, 8, 0.9))
row3policy = (valueiteration(row3states, goal3, row3mdp, 16, 0.9))
```

```python
def nextstate(state, act):
    next_state = list(state).copy()
    hole = next_state.index(16)
    if act=='l':
        if hole%4 == 0:
            return state
        next_state[hole], next_state[hole-1] = next_state[hole-1], next_state[hole]
    elif act=='r':
        if hole%4 == 3:
            return state
        next_state[hole], next_state[hole+1] = next_state[hole+1], next_state[hole]
    elif act=='u':
        if hole//4 == 0:
            return state
        next_state[hole], next_state[hole-4] = next_state[hole-4], next_state[hole]
    elif act=='d':
        if hole//4 == 3:
            return state
        next_state[hole], next_state[hole+4] = next_state[hole+4], next_state[hole]
    return tuple(next_state)

def convertor(state, num):
    st = list(state).copy()
    for i in range(16):
        if st[i]>num and st[i]!=16:
            st[i]=0
    return tuple(st)

def out(state):
    l = [3,7,11, 15]
    for i in range(16):
        print(state[i], end=' ')
        if i in l:
            print()
```
